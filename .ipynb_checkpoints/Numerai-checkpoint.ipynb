```python
eras = np.unique(training_data['era'])
```

```python
feature_cols = training_data.columns[training_data.columns.str.startswith('feature')]
```

```python
live_data = tournament_data[tournament_data['data_type'] == 'live']
```

```python
rf = RandomForestRegressor(max_depth = 8, n_estimators = 300, n_jobs = -1, oob_score=True)
```

```python
len(eras)
```

```python
predictions = []
oobs = []
rfs = []
for era in eras:
    era_df = training_data[training_data['era'] == era]
    X = era_df[feature_cols]
    y = era_df['target_kazutsugi']
    rf.fit(X, np.log(y + 1e-6))
    oobs.append(rf.oob_score_)
    rfs.append(rf)
    predictions.append(rf.predict(live_data[feature_cols]))
```

```python
sorted(list(zip(feature_cols, rf.feature_importances_)), key = lambda x: x[1], reverse = True)
```

```python
predictions = np.array(predictions)
```

```python
predictions.max()
```

```python
predictions.min()
```

```python
predictions.shape
```

```python
kmeans = KMeans(n_clusters =2)
```

```python
kmeans.fit(predictions)
```

```python
groups = kmeans.predict(predictions)
```

```python
silhouette_score(predictions, groups)
```

```python
hi_score = 0
for i in range(2, 101):
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(predictions)
    groups = kmeans.predict(predictions)
    score = silhouette_score(predictions, groups)
    if score > hi_score:
        hi_score = score
        best_clusters = i
```
